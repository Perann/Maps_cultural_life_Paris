# Statistiques sur les concerts de musiqiue classique à Paris 

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/home/onyxia/work/Maps_cultural_life_Paris/Outputs/DataSets/DataMusiqueClassique_v8.csv')
```

# Combien de concerts de musique classique sont actuellment programmés 

```python
print("{} concerts de musique classique sont actuellement programmés à Paris entre Décembre et Mai".format(len(df)))
```

# Où sont les concerts de musique classique à Paris 

```python
arrondissements_df = df.groupby('commune')['nom'].count().reset_index()

arrondissements_df = arrondissements_df.rename(columns={'nom': 'nombre_de_concerts'})

arrondissements_df = arrondissements_df.sort_values(by = 'nombre_de_concerts', ascending = False)

arrondissements_df.head(3)
```

```python
bar_1 = arrondissements_df.head(20).plot(x = 'commune', y = 'nombre_de_concerts',kind = 'bar')

for i, value in enumerate(arrondissements_df.head(20)['nombre_de_concerts']):
    bar_1.text(i, value + 1, str(value), ha='center', va='bottom')

plt.show()
```

# Quelles sont les salles qui proposent le plus de concerts

```python
salles_df = df.groupby('etablissement')['nom'].count().reset_index()
salles_df = salles_df.rename(columns={'nom': 'nombre_de_concerts'})
salles_df = salles_df.sort_values(by = 'nombre_de_concerts', ascending = False)
salles_df.head(5)
```

```python
salles_df.tail(5)
```

```python
print("Il y a {} salles qui proposent des concerts de musique classique en ce moment à Paris".format(len(salles_df)))
```

```python
nombre_eglises = salles_df['etablissement'].str.lower().str.count('église'.lower())
total_eglises = nombre_eglises.sum()

print(f"Nombre total d'églises : {total_eglises}")
```

```python
nombre_philharmonie = salles_df['etablissement'].str.lower().str.count('Philharmonie'.lower())
total_philharmonie_salles = nombre_philharmonie.sum()

print(f"Nombre de salles de la Philharmonie : {total_philharmonie_salles}")
```

```python
top20_salles_df = salles_df.head(20)

top20_salles_df.plot(x = 'etablissement',y = 'nombre_de_concerts',kind = 'bar')
```

```python
salles_philharmonie = salles_df[salles_df['etablissement'].str.contains('philharmonie', case=False, na=False)]
total_philharmonie_concerts = salles_philharmonie['nombre_de_concerts'].sum()

total_philharmonie = pd.DataFrame({'etablissement': ['Total Philharmonie'], 'nombre_de_concerts': [total_philharmonie_concerts]})

salles_v2_df = pd.concat([total_philharmonie, salles_df], ignore_index=True)
salles_v2_df.head()
```

```python
top20_salles_v2_df = salles_v2_df.head(20)

# Création du graphique
ax = top20_salles_v2_df.plot(x='etablissement', y='nombre_de_concerts', kind='bar', legend=False)

# Changer la couleur de la première barre en rouge
ax.patches[0].set_facecolor('red')

# Ajouter les quantités au-dessus des barres
for i, value in enumerate(top20_salles_v2_df['nombre_de_concerts']):
    ax.text(i, value + 1, str(value), ha='center', va='bottom')

# Affichage du graphique
plt.show()
```

# Combien y a t il de concerts par mois ? 

```python
df['date datetime'] = pd.to_datetime(df['date datetime'])
concerts_par_mois = df.groupby(df['date datetime'].dt.to_period('M')).size().reset_index(name='nombre de concerts')

concerts_par_mois
```

```python
concerts_par_mois['date numeric'] = concerts_par_mois['date datetime'].dt.to_timestamp().astype(int)


plt.figure(figsize=(8, 8))
plt.plot(concerts_par_mois['date numeric'], concerts_par_mois['nombre de concerts'], marker='o', linestyle='-', color='b')


plt.title('Nombre de Concerts par Mois à date', fontsize=16)
plt.xlabel('Mois', fontsize=14)
plt.ylabel('Nombre de Concerts', fontsize=14)


plt.xticks(concerts_par_mois['date numeric'], [period.strftime('%b %Y') for period in concerts_par_mois['date datetime']], rotation=45, fontsize=14)
plt.yticks(fontsize=14)


for i, value in enumerate(concerts_par_mois['nombre de concerts']):
    plt.text(
        concerts_par_mois['date numeric'][i], 
        value + 2, 
        str(value), 
        ha='center', 
        fontsize=14,
        bbox=dict(facecolor='white', alpha=0.7, edgecolor='white')
    )


plt.show()
```
