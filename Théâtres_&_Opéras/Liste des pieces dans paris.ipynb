```python
!pip install -q lxml

import bs4
import lxml
import pandas as pd
import urllib

from urllib import request

url_liste_etablissements = "https://www.offi.fr/theatre/pieces-de-theatre.html?criterion_sch_ville=75"
    
request_text = request.urlopen(url_liste_etablissements).read()
type(request_text)

page = bs4.BeautifulSoup(request_text, "lxml")
```

```python
balises_b = page.find_all('b')
for balise in balises_b:
    print(balise.text)

balises_p = page.find_all('p')
for balise in balises_p:
    print(balise.text)

lignes_with_span_name = page.find_all('span', {'itemprop': 'name'})
contenu_balises = [ligne.text for ligne in lignes_with_span_name]
for contenu in contenu_balises:
    print(contenu)



balises_a = page.find_all('a', class_='text-body')
for balise_a in balises_a:
    texte_a = balise_a.text.strip()
    print(texte_a)
#### 
# Extract text from <b> tags
balises_b = page.find_all('b')
text_from_b = [balise.text.strip() for balise in balises_b]

# Extract text from <p> tags
balises_p = page.find_all('p')
text_from_p = [balise.text.strip() if balise else None for balise in balises_p]

# Extract text from <span itemprop="name"> tags
lignes_with_span_name = page.find_all('span', {'itemprop': 'name'})
text_from_span_name = [ligne.text.strip() for ligne in lignes_with_span_name]

# Extract text from <a> tags with class 'text-body'
balises_a = page.find_all('a', class_='text-body')
text_from_a = [balise_a.text.strip() for balise_a in balises_a]

# Determine the maximum length among the lists
max_length = max(len(text_from_b), len(text_from_p), len(text_from_span_name), len(text_from_a))

# Pad the lists with None to make them of equal length
text_from_b += [None] * (max_length - len(text_from_b))
text_from_p += [None] * (max_length - len(text_from_p))
text_from_span_name += [None] * (max_length - len(text_from_span_name))
text_from_a += [None] * (max_length - len(text_from_a))

# Create a DataFrame
df = pd.DataFrame({
    'Text from <b> tags': text_from_b,
    'Text from <p> tags': text_from_p,
    'Text from <span itemprop="name"> tags': text_from_span_name,
    'Text from <a> tags': text_from_a,
})

# Display the DataFrame
print(df)
```

```python
divs_with_class = page.find_all('div', class_='has-padding-left-20 has-padding-right-20 has-padding-bottom-20')

# Extraire le texte à l'intérieur de la balise <b> pour chaque div
infos = [div.find('b').text.strip() for div in divs_with_class]

# Afficher les informations extraites
for info in infos:
    print(info)
```

```python
paragraphes_with_class = page.find_all('p', class_='card-text mb-auto')

# Extraire le texte pour chaque paragraphe
infos = [paragraphe.text.strip() for paragraphe in paragraphes_with_class]

# Afficher les informations extraites
for info in infos:
    print(info)
```

```python
name_with_class = page.find_all('a', class_='text-body')

# Extraire le texte pour chaque paragraphe
infos = [name.text.strip() for name in name_with_class]

# Afficher les informations extraites
for info in infos:
    print(info)
```
